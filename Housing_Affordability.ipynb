```python
#feature engineering
master = master.sort_values(['City', 'Year'])

master['Population_Growth'] = (
    master.groupby('City')['Population']
    .pct_change() * 100
)
```

```python
master['HPI_Growth'] = (

master.groupby('City')
['Housing_Price_Index']

.pct_change()

*100

)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
#creates the scaled variables
master[['Rent_scaled',
        'HPI_scaled',
        'Pop_scaled',
        'Vacancy_scaled']] = scaler.fit_transform(

master[['Rent_2025',
        'Housing_Price_Index',
        'Population_Growth',
        'Vacancy_2025']]
)
```

```python
#equal weights
master['Housing_Pressure_Score'] = (

    0.25 * master['Rent_scaled']

    + 0.25 * master['HPI_scaled']

    + 0.25 * master['Pop_scaled']

    - 0.25 * master['Vacancy_scaled']

)
```
